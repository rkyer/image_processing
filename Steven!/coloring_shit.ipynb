```python
%matplotlib inline

import numpy as np
import matplotlib.pylab as plt
import astropy.io.fits as fits

from glob import glob

import os

from os.path import expanduser

from astropy.visualization import make_lupton_rgb
from matplotlib.colors import LogNorm
```

```python
home = expanduser('~')
path = os.path.join(home,'Downloads','h_m51**.fits')
filepath = glob(path, recursive=True)
```

```python
filtkeyzx = ['halpha?','V_probably','B_probably','i']
```

```python
filts = []
for file in filepath:
    filts.append(fits.getdata(file))
```

```python
plt.imshow(filts[0],norm=LogNorm(), cmap='rainbow')
plt.gca().invert_yaxis();
```

RGB bitches:

```python
pretty_ass_picture = make_lupton_rgb(5*filts[3], 10*filts[1], 10*filts[2], stretch=5)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(pretty_ass_picture, norm=LogNorm(vmin=200), cmap='rainbow')
plt.gca().invert_yaxis();
```
