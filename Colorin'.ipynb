```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from astropy.io import fits
import astropy.units as u
import glob
import astropy
```

```python
def plot_image(a,figsize=(10,10), xlabel=None,ylabel=None,title=None, cmap='binary_r',alpha=1):
    plt.figure(figsize=figsize)
    plt.imshow(a, vmin=np.percentile(a, 5), vmax=np.percentile(a, 98),
               origin='lower', cmap=cmap, interpolation='nearest',alpha=alpha)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
```

```python
fnms = glob.glob("*.fits")
data = {fnms[i]:0 for i in range(len(fnms))}
for i in fnms:
    data[i] = fits.getdata(i)
#     data[i] /= np.mean(data[i])
```

```python
from astropy.visualization import make_lupton_rgb
plt.figure(figsize=(10,10))
image = make_lupton_rgb(data[fnms[2]], 2*data[fnms[3]], 2*data[fnms[0]], stretch=0.8)
plt.imshow(image)
```

```python
a = image
plt.figure(figsize=(10,10))
plt.imshow(a, vmin=np.percentile(a, 5), vmax=np.percentile(a, 98),
               origin='lower', interpolation='nearest',alpha=1)
a = 50*data[fnms[2]]
plt.imshow(a, vmin=np.percentile(a, 5), vmax=np.percentile(a, 98),
               origin='lower',cmap='PuRd', interpolation='nearest',alpha=.15)
plt.show()
a = image
plt.figure(figsize=(10,10))
plt.imshow(a, vmin=np.percentile(a, 5), vmax=np.percentile(a, 98),
               origin='lower', interpolation='nearest',alpha=1)
plt.show();
```
